```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('Social_Network_Ads.csv')
```

```python
df.sample(5)
```

```python
df.info()
```

```python
df.describe()
```

```python
df.shape
```

```python
x = df.drop('Purchased', axis=1)
y = df['Purchased']
```

```python
x.head()
```

```python
y.head()
```

```python
x.mean()
```

```python
y.mean()
```

```python
x.std()
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.3,random_state= 0)
x_train.head()
```

```python
x_train.shape
```

```python
x_test.shape
```

```python
x_test.head()
```

```python
y_train.head()
```

```python
y_test.head()
```

```python
y_test.shape
```

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
sc.fit(x_train)

x_train_scaled = sc.transform(x_train)
x_test_scaled = sc.transform(x_test)
```

```python
x_train_scaled = pd.DataFrame(x_train_scaled,columns = ['Age', 'EstimatedSalary'])
x_test_scaled = pd.DataFrame(x_test_scaled,columns = ['Age', 'EstimatedSalary'])
```

```python
x_train_scaled.describe()
```

```python
fig , (ax1,ax2) = plt.subplots(1,2,figsize=(10,5))
ax1.scatter(x='Age', y='EstimatedSalary', data=x_train_scaled)
ax2.scatter(x='Age', y='EstimatedSalary', data=x_train,color='red')
plt.show()
```

# Medical Dataset

```python
df = pd.read_csv('Medicaldataset.csv')
df.sample(5)
```

```python
df.info()
```

```python
df.shape
```

```python
df.describe()
```

```python
x = df.drop(['Age','Gender','Blood sugar','CK-MB','Troponin','Result'],axis = 1)
y = df['Result']
```

```python
x.head()
```

```python
y.head()
```

```python
x.shape,y.shape
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.3,random_state=0)
```

```python
x_train.shape,x_test.shape
```

```python
y_train.shape,y_test.shape
```

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()

sc.fit(x_train)

x_train_scaled = sc.transform(x_train)
x_test_scaled = sc.transform(x_test)
```

```python
x_train_scaled = pd.DataFrame(x_train_scaled,columns = ['Heart rate',	'Systolic blood pressure','Diastolic blood pressure'])
x_test_scaled = pd.DataFrame(x_test_scaled,columns = ['Heart rate',	'Systolic blood pressure','Diastolic blood pressure'])
```

```python
np.round(x_train_scaled.describe())
```

```python
sns.scatterplot(x='Heart rate',y='Systolic blood pressure',data=x_train)
```

```python
sns.scatterplot(x='Heart rate',y='Systolic blood pressure',data=x_train_scaled)
```

```python
fig, (ax1,ax2,ax3) = plt.subplots(1,3,figsize=(10,5))
ax1.scatter(x='Heart rate',y='Systolic blood pressure',data=x_train)
ax2.scatter(x='Heart rate',y='Systolic blood pressure',data=x_train_scaled,color = 'red')
ax3.scatter(x='Heart rate',y='Diastolic blood pressure',data=x_train_scaled,color = 'red')
plt.show()
```
